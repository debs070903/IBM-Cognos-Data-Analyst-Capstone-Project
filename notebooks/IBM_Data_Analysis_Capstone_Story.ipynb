```python
import pandas as pd

df = pd.read_csv("survey_data_updated 5.csv")
df.head()
```

```python
cols = [
    "ResponseId",
    "Country",
    "Age",
    "EdLevel",

    "LanguageHaveWorkedWith",
    "DatabaseHaveWorkedWith",
    "PlatformHaveWorkedWith",
    "WebframeHaveWorkedWith",

    "LanguageWantToWorkWith",
    "DatabaseWantToWorkWith",
    "PlatformWantToWorkWith",
    "WebframeWantToWorkWith"
]

df = df[cols]
df.head()
```

```python
df = df.drop_duplicates()
df.replace("", pd.NA, inplace=True)
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
multi_cols = [
    "LanguageHaveWorkedWith",
    "DatabaseHaveWorkedWith",
    "PlatformHaveWorkedWith",
    "WebframeHaveWorkedWith",
    "LanguageWantToWorkWith",
    "DatabaseWantToWorkWith",
    "PlatformWantToWorkWith",
    "WebframeWantToWorkWith"
]

for col in multi_cols:
    df[col] = df[col].dropna().str.split(";")
```

```python
lang_current = df[["ResponseId", "LanguageHaveWorkedWith"]].explode(
    "LanguageHaveWorkedWith"
)

lang_current = lang_current.dropna().reset_index(drop=True)
lang_current.columns = ["Respondent", "Language"]
```

```python
db_current = df[["ResponseId", "DatabaseHaveWorkedWith"]].explode(
    "DatabaseHaveWorkedWith"
)

db_current = db_current.dropna().reset_index(drop=True)
db_current.columns = ["Respondent", "Database"]
```

```python
platform_current = df[["ResponseId", "PlatformHaveWorkedWith"]].explode(
    "PlatformHaveWorkedWith"
)

platform_current = platform_current.dropna().reset_index(drop=True)
platform_current.columns = ["Respondent", "Platform"]
```

```python
web_current = df[["ResponseId", "WebframeHaveWorkedWith"]].explode(
    "WebframeHaveWorkedWith"
)

web_current = web_current.dropna().reset_index(drop=True)
web_current.columns = ["Respondent", "WebFramework"]
```

```python
lang_future = df[["ResponseId", "LanguageWantToWorkWith"]].explode(
    "LanguageWantToWorkWith"
)

lang_future = lang_future.dropna().reset_index(drop=True)
lang_future.columns = ["Respondent", "Language"]
```

```python
db_future = df[["ResponseId", "DatabaseWantToWorkWith"]].explode(
    "DatabaseWantToWorkWith"
)

db_future = db_future.dropna().reset_index(drop=True)
db_future.columns = ["Respondent", "Database"]
```

```python
platform_future = df[["ResponseId", "PlatformWantToWorkWith"]].explode(
    "PlatformWantToWorkWith"
)

platform_future = platform_future.dropna().reset_index(drop=True)
platform_future.columns = ["Respondent", "Platform"]
```

```python
web_future = df[["ResponseId", "WebframeWantToWorkWith"]].explode(
    "WebframeWantToWorkWith"
)

web_future = web_future.dropna().reset_index(drop=True)
web_future.columns = ["Respondent", "WebFramework"]
```

```python
demographics = df[["ResponseId", "Country", "Age", "EdLevel"]].dropna(
    how="all"
)

demographics.columns = ["Respondent", "Country", "AgeGroup", "EducationLevel"]
```

```python
lang_current.to_csv("current_languages.csv", index=False)
db_current.to_csv("current_databases.csv", index=False)
platform_current.to_csv("current_platforms.csv", index=False)
web_current.to_csv("current_webframeworks.csv", index=False)

lang_future.to_csv("future_languages.csv", index=False)
db_future.to_csv("future_databases.csv", index=False)
platform_future.to_csv("future_platforms.csv", index=False)
web_future.to_csv("future_webframeworks.csv", index=False)

demographics.to_csv("demographics.csv", index=False)
```

```python
import matplotlib.pyplot as plt

lang_current = pd.read_csv("current_languages.csv")

top10_lang_current = (
    lang_current
    .value_counts("Language")
    .head(10)
    .reset_index()
)

top10_lang_current.columns = ["Language", "Count"]

plt.figure()
plt.bar(top10_lang_current["Language"], top10_lang_current["Count"])
plt.title("Top 10 Programming Languages (Current Year)")
plt.xlabel("Programming Language")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
lang_current = pd.read_csv("current_languages.csv")
lang_future = pd.read_csv("future_languages.csv")
platform_current = pd.read_csv("current_platforms.csv")
demographics = pd.read_csv("demographics.csv")

curr_lang = lang_current.value_counts("Language").head(10)
fut_lang = lang_future.value_counts("Language").head(10)

lang_compare = pd.DataFrame({
    "Current Year": curr_lang,
    "Next Year": fut_lang
}).fillna(0)

top_platforms = platform_current.value_counts("Platform").head(10)

top_countries = demographics.value_counts("Country").head(10)

age_edu = (
    demographics
    .groupby(["AgeGroup", "EducationLevel"])
    .size()
    .unstack(fill_value=0)
)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

lang_compare.plot(kind="bar", ax=axes[0, 0])
axes[0, 0].set_title("Top Programming Languages: Current vs Next Year")
axes[0, 0].set_xlabel("Language")
axes[0, 0].set_ylabel("Respondent Count")
axes[0, 0].tick_params(axis="x", rotation=45)

axes[0, 1].barh(top_platforms.index, top_platforms.values)
axes[0, 1].set_title("Top 10 Platforms Worked With (Current Year)")
axes[0, 1].set_xlabel("Respondent Count")
axes[0, 1].set_ylabel("Platform")

axes[1, 0].bar(top_countries.index, top_countries.values)
axes[1, 0].set_title("Top 10 Countries by Respondents")
axes[1, 0].set_xlabel("Country")
axes[1, 0].set_ylabel("Respondent Count")
axes[1, 0].tick_params(axis="x", rotation=45)

age_edu.plot(kind="bar", stacked=True, ax=axes[1, 1])
axes[1, 1].set_title("Respondents by Age Group and Education Level")
axes[1, 1].set_xlabel("Age Group")
axes[1, 1].set_ylabel("Respondent Count")
axes[1, 1].tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.savefig("additional_analysis_combined.png", dpi=300, bbox_inches="tight")
plt.show()
```
